```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("data/netflix_mini.csv")
df.head()
```

```python
print("记录总数：",len(df))
print("电影数量：",sum(df['type']=="Movie"))
print("剧集数量：",sum(df['type']=="TV Show"))
print("发行年份：\n",df['release_year'].value_counts().sort_index())
```

```python
sns.countplot(data=df,x="type",palette="Set2") #Set2柔和配色
plt.title("Movies vs TV Shows")
plt.show()
```

```python
#发行年份趋势
df["release_year"].value_counts().sort_index().plot(kind="line",marker="o")
#类型是折线图，数据在图中用圆点表示
plt.title("发行年份趋势")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
#前5国家
df['country'].value_counts().head(5).plot(kind="bar")
plt.title("TOP 5")
plt.ylabel("Count")
plt.show()
```

```python
#导出清洗后的数据
df.to_csv("netflix_cleaned.csv",index=False)
print("已保存为netflix_cleaned.csv")
```
